```python
import numpy as np
import pandas as pd
```

```python
dic={
    "Name":["Hari","Krishna","Manoj","Akhil","Akhil"],
    "Age":[18,180,19,20,np.nan],
    "Salary":[100000,200000,35000,75000,20000]
}
df=pd.DataFrame(dic)
df
```

```python
#Missing values
df_droped=df.dropna()
print("Dataset After droping null values")
df_droped
```

```python
m=df['Age'].median()
m=int(m)
df['Age']=df['Age'].fillna(m)
print("Imputation in missing values")
df
```

```python
df=df.sort_values(by='Age')
q1=df['Age'].quantile(0.25)
q3=df['Age'].quantile(0.75)
IQR=q3-q1
l=q1-1.5*IQR
u=q3+1.5*IQR
Clen_df=df[(df['Age']>=l) & (df['Age']<=u)]
print("dTASET aFTER REMOVING TRTHE oYTLIERS")
Clen_df
```

```python
final_df=Clen_df.drop_duplicates(keep='first')
print("Final cleaned Dataset is")
final_df.dropna(inplace=True)
final_df
```
